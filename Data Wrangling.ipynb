### 2.Write functions that will (1) convert the answer on each of the personality test questions to a number, as set out in scoring, and (2) calculate total scores for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion).
```python
def assigning(row,Category_Num):
    count = 0
    asd = 0
    for column in personality_data.columns:
        tuplee = row[column]
        if type(tuplee) == str:
            catergory = tuplee[1]
            result = int(tuplee[4])
            if catergory == Category_Num:
                count += result
    return count
```

```python
personality_data['Extraversion'] = personality_data.apply(lambda row : assigning(row,'1'), axis = 1)
personality_data['Agreeableness'] = personality_data.apply(lambda row : assigning(row,'2'), axis = 1)
personality_data['Conscientiousness'] = personality_data.apply(lambda row : assigning(row,'3'), axis = 1)
personality_data['Emotional Stability'] = personality_data.apply(lambda row : assigning(row,'4'), axis = 1)
personality_data['Openness to experiences'] = personality_data.apply(lambda row : assigning(row,'5'), axis = 1)
```

### 1 . Create a function/functions that takes in subscale total, and for each subscale, creates a new column called “{subscale_name}_interpretation” (e.g., conscientiousness_interpretation) and applies the labels “low”, “medium”, or “high” according to a person’s score on that subscale. Cut-off scores corresponding to the low, medium and high groups can be found in scoring.md. Again, write a unit test before creating this function that tests whether the generate_score_interpretation function works.

```python
def subscale(score,name):
    mean = personality_data[name].mean()
    std = personality_data[name].std()
    high = (mean + std)/2
    low = (mean - std)/2
    
    if score >= high:
        return 'high'
    elif score <= low:
        return 'low'
    else:
        return 'medium'
```

```python
personality_data['Emotional_Stability_Interpretation'] = personality_data['Emotional Stability'].apply(lambda column: subscale(column,'Emotional Stability'))
personality_data['Agreeableness_Interpretation'] = personality_data['Agreeableness'].apply(lambda column: subscale(column,'Agreeableness'))
personality_data['Conscientiousness_Interpretation'] = personality_data['Conscientiousness'].apply(lambda column: subscale(column,'Conscientiousness'))
personality_data['Emotional_Stability_Interpretation'] = personality_data['Emotional Stability'].apply(lambda column: subscale(column,'Emotional Stability'))
personality_data['Openness_to_experiences_Interpretation'] = personality_data['Openness to experiences'].apply(lambda column: subscale(column,'Openness to experiences'))
```
